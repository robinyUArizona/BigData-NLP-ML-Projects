```
## Import libraries
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding, Activation, Dropout
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D 

import numpy as np
from numpy import array
import pandas as pd
```

```
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('twitter4000.csv')
df.head()
```

```
df['sentiment'].value_counts(dropna = False)
```

```
text = df['tweets'].tolist()
text[:20]
```

```
y = df['sentiment']
```

```
token = Tokenizer()
token.fit_on_texts(text)
```

```
token
```

```
## Dictionary
print(token.index_word)
```

```
## Total no. of unique words
vocab_size = len(token.word_index) + 1
```

```
vocab_size # Total no. of unique words
```

```
# token.index_word
```

```
## Example of encoding
x = ['i to the a and']
# x = [1, 2, 3, 4, 6]
```

```
token.texts_to_sequences(x)
```

```
## We just need index (keys) of words
encoded_text = token.texts_to_sequences(text)
print(encoded_text)
```

```
## Note from above out put
## -- Each vector above might have different size
## -- So therefore we need pad_sequence

max_length = 120 ## All tweets can have maximum words of 120 ---LOL
X = pad_sequences(encoded_text, maxlen = max_length, padding = 'post')
```

```
print(X)
```

```
X.shape
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    random_state = 42, 
                                                    test_size = 0.2,
                                                    stratify = y)
```

```
vec_size = 300

model = Sequential()
model.add(Embedding(vocab_size, vec_size, input_length = max_length))

## convolution layer
model.add(Conv1D(64, 8, activation = 'relu'))
model.add(MaxPooling1D(2)) # with 2 data
model.add(Dropout(0.5))

model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(16, activation = 'relu'))

model.add(GlobalMaxPooling1D())


model.add(Dense(1, activation = 'sigmoid'))

```

```
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
%%time
model.fit(X_train, y_train, epochs = 5, valdiation_data = (X_test, y_test))
```

```
def get_encoded(x):
    x = token.texts_to_sequences(x)
    x = pad_sequences(x, maxlen = max_length, padding='post')
    return x
```

```
x = [' worst services. will not not come again']
```

```
model.predict_classes(get_encoded(x))
```
